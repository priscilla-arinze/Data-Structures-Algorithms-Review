# Data Structure - Stacks

LIFO - Last In, First Out (removes the item that has been in the list for the shortest amount of time)

<hr>

__*Can be demonstrated with lists/arrays or `deque()` from the collections library*__
* `deque()` can avoid inefficient/extra memory usage from Python dynamic lists
* `deque()` uses a doubly linked list
* `deque()` is pronounced like "deck" and is short for "double-ended queue"



__Stacks can be great for:__
* navigating web browsing history with forward and back buttons
* navigating action history with undo (`Ctrl + Z`) & redo (`Ctrl + Y`)


*Major Fun Fact Alert: Anytime a program is executed in any programming language, each operation and/or function invocation is sent to the "call stack" using the system's RAM & is executed in the order of the call stack*
* Can be seen with an IDE debugger

### `append()` *(`push()`)*: Adds items to a list in a stack-like fashion --- *O(1)*

*Very first item is the item at the very bottom or left*

*Very last item is the item at the very top or right*

```python
books = []

books.append("book 1") ## BOTTOM of stack
books.append("book 2")
books.append("book 3")
books.append("book 4") ## TOP of stack

books
```

### `pop()`: Retrieves last added item & takes it out of list --- *O(1)* 

```python
last_book = books.pop()
last_book
```

```python
books
```

```python
# pop() again
last_book2 = books.pop()
last_book2
```

```python
books
```

### `list[-1]` *(`peek()`)*: Retrieves last added item, but does NOT remove it from list --- *O(1)*

__OR: `len(list)` - 1__

```python
books[-1]
```

```python
books
```

## Custom mini Stack classes for `deque()` & `list()`

```python
class Stack_Deque:
    def __init__(self):
        self.stack = deque() # create a new deque() object upon class object creation

    # Add items to stack
    def push(self, data):
        return self.stack.append(data)

    # Retrieve last added item & remove from deque object
    def pop(self):
        return self.stack.pop()
    
    # Retrieve last added item, but DO NOT remove from deque object
    def peek(self):
        return self.stack[-1]

    # Check to see if deque object is empty; returns a boolean of True if it is empty
    def is_empty(self):
        return len(self.stack) == 0

    # Returns the number of items in the deque object
    def size(self):
        return len(self.stack)

    # What to print out when Stack_Deque object name is invoked
    def __repr__(self):
        if self.is_empty():
            return "No items have been added to deque yet"
        
        else:
            items = []
            reverse_index = self.size() - 1
            max_index = self.size() - 1

        while reverse_index >= 0:
            # First item added
            if reverse_index == 0:
                items.append("[FIRST: %s]" % self.stack[reverse_index])
            
            # Last item added
            elif reverse_index == max_index:
                items.append("[LAST: %s]" % self.stack[max_index])

            # All the rest of the items
            else:
                items.append("[%s]" % self.stack[reverse_index])


            reverse_index -= 1
        
        
        return '\n'.join(items)




class Stack_List:
    def __init__(self):
        self.stack = [] # create a new list object upon class object creation

    # Add items to stack
    def push(self, data):
        return self.stack.append(data)

    # Retrieve last added item & remove from list
    def pop(self):
        return self.stack.pop()
    
    # Retrieve last added item, but DO NOT remove from list
    def peek(self):
        return self.stack[-1]

    # Check to see if list is empty; returns a boolean of True if it is empty
    def is_empty(self):
        return len(self.stack) == 0

    # Returns the number of items in the list
    def size(self):
        return len(self.stack)

    # What to print out when Stack_List object name is invoked
    def __repr__(self):
        if self.is_empty():
            return "No items have been added to list yet"
        
        else:
            items = []
            reverse_index = self.size() - 1
            max_index = self.size() - 1

        while reverse_index >= 0:
            # First item added
            if reverse_index == 0:
                items.append("[FIRST: %s]" % self.stack[reverse_index])
            
            # Last item added
            elif reverse_index == max_index:
                items.append("[LAST: %s]" % self.stack[max_index])

            # All the rest of the items
            else:
                items.append("[%s]" % self.stack[reverse_index])


            reverse_index -= 1
        
        
        return '\n'.join(items)
```

```python
states = Stack_Deque()

states.push("california") # bottom of stack
states.push("texas")
states.push("delaware")
states.push("florida")
states.push("ohio") # top of stack

states
```
