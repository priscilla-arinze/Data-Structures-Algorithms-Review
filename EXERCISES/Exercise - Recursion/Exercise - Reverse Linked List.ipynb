### Iterative Approach

```python
def iterative_reverse_linkedlist(linked_list):
    if linked_list == None or linked_list.head == None or linked_list.size() == 1:
        return linked_list

    reversed_linked_list = LinkedList()
    reversed_index = linked_list.size() - 1


    while reversed_index >= 0:
        reversed_linked_list.append_item(linked_list.get_node_at_index(reversed_index).data)

        reversed_index -= 1

    return reversed_linked_list
```

```python
ls = LinkedList()

ls.append_item(10)
ls.append_item(20)
ls.append_item(30)
ls.append_item(40)
ls.append_item(50)
ls.append_item(60)

ls
```

```python
reversed_ls = iterative_reverse_linkedlist(ls)

reversed_ls
```

### Recursive Approach

*Heavily used VS Code Debugging Tool in a separate python file*

```python
def reverse(linked_list, node):
    if node == None or node.next_node == None:
        return node
    
    # Will execute this line until it reaches the last recursion, sets the final recursion value to "temp_tail" (last node in original list)
    temp_tail = reverse(linked_list, node.next_node)

    # NOTE: node variable is set to the node before the very last node ("temp tail" variable)

    # Update the current node's pointer to 2 nodes ahead 
    # e.g. move 4 at the end, [1->-2->3->4->5-X] ---> [5-X 1->2->3->4]; see visualization below)
    node.next_node.next_node = node

    # IMPORTANT: update the linkedlist's head value to the last node in original list 
    linked_list.head = temp_tail

    # To avoid a circular/double pointer reference after updating current node pointer above, change the next node of current node to None
    # Will be updated to a new pointer as it goes through the remaining recursion calls in the call stack
    node.next_node = None




    print("Reversed Linked List:", linked_list)

    
    # Keeps returning last node in original list
    return temp_tail
```

```python
def verify_reversed_linked_list(node):
    tmp = node
    print("Verifying reversal...:", end = " ")

    while tmp != None:
        print(tmp.data, end=" -> ")

        tmp = tmp.next_node
```
