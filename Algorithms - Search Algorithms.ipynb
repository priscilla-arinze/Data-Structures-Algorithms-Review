# Search Algorithms


### Linear/sequential search --- *O(n)*
* If you need to find a specific number in a list of 1000 numbers:
    * you will need to start with the first number as the biggest number
    * go through each number and check to see if it is bigger than prev number 
    * repeat until you reach the last number

### Binary search (or "Divide & Conquer") --- *O(log n)*
* Essentially starts at the middle of the list & repeatedly divide the sub-lists in half to get to target number/item
* If you need to find a specific number in a list of 1000 numbers:
    * make sure to have a sorted array
    * divide the list in half & get the middle number
    * check to see if the target number is bigger or lower than the middle number, then only focus on the right or left side of the divided list, respectively
    * Repeat for each subsection until you reach the target number
* Works best for __*sorted*__ lists
